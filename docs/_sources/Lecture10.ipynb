# Lecture 10 - Torque and rotational acceleration
We are almost at the end of our study of rotational kinematics, but there are still two topics that we have only covered in the linear kinematics section. These as the rotational equivalent of force known as **torque** and rotational equivalent of momentum known, unsurprisingly, as **angular momentum**.

## Torque
### Definition of torque
When we considered Newton's Second Law we stated that a non-zero net force will result in some acceleration (assuming the mass remains constant). Although this is true it is not the whole picture. It is entirely feasbile that a body experiencing some external force may not accelerate in a linear motion but instead begins to rotate. In this case we are concerning ourselves with the force acting in a direction perpendicular to the line connecting the pivot point and the point (or points) on the object where the force(s) are being applied. Let us consider a diagram of a bar of length $r$ with a fixed pivot point at one end and a force applied to the other end ({numref}`fig:C10_torque`).

```{figure} Figures/C10_torque.png
:name: fig:C10_torque
:alt: A force $F$ is applied to the end of a bar of length $r$ which is fixed to a pivot point at the opposite end. This force causes the bar to rotate.
:width: 70%
:align: center
A force $F$ is applied to the end of a bar of length $r$ which is fixed to a pivot point at the opposite end. This force causes the bar to rotate.
```

````{margin}
```{note}
You may need to cast your memory back to school where you studied levers. If you have two forces acting either side of a pivot point then the leverage, i.e force times distance, need to balance if you want the lever / seesaw to balance.
```
````
We assume, for the sake of generality, that the force $F$ is acting in some direction between $\hat{\textbf{r}}$ (i.e. pulling on the bar) and $\hat{\boldsymbol{\theta}}$ (i.e. only acting to induce rotation). It is the component of the force acting perpendicular to the length of that bar that will result in some angular acceleration of the bar, and this component is $F_\text{rot} = F\sin\phi$. Therefore the leverage or **torque** $\boldsymbol{\tau}$ on the bar is
\begin{align*}
\boldsymbol{\tau} &= rF\sin\phi\,\hat{\boldsymbol{\theta}}\\
&= \boldsymbol{r}\times\boldsymbol{F}
\end{align*}

If we have multiple forces acting on the body then the total torque is equal to the sum of the individual torques, i.e.
\begin{align*}
\boldsymbol{\tau} &=\sum_i \left(\boldsymbol{r}_i\times\boldsymbol{F}_i\right)
\end{align*}

### Angular acceleration and torque
Any net force will result in an acceleration, as we know from Newton's Second Law. What we are currently missing is a direct relationship between the torque and resulting angular acceleration, in the same way that we know the relationship between force and acceleration along a straight line.

We can consider a solid body of mass $m$ rotating about some axis as the sum of many many small mass elements $\mathrm{d}m$. We have already used this idea when deriving the [total energy of a rotating body](sec:C8_energy) (specifically {numref}`fig:C8_energy`), but now we are applying some small force to each of these masses such that a tangential force $\mathrm{d}F_t$ is applied as shown in {numref}`fig:C10_accel`.

```{figure} Figures/C10_accel.png
:name: fig:C10_accel
:alt: A force is applied to a random shaped mass that causes it to rotate about an axis perpendicular to the plane of the diagram. The mass is split into small mass elements each of which has a tangential force $\mathrm{d}F_t$ acting on it.
:width: 70%
:align: center
A force is applied to a random shaped mass that causes it to rotate about an axis perpendicular to the plane of the diagram. The mass is split into small mass elements each of which has a tangential force $\mathrm{d}F_t$ acting on it.
```

We use Newton's Second law to state
\begin{align*}
\mathrm{d}F_t &= a_t\,\mathrm{d}t
\end{align*}
where $a_t$ is the tangential acceleration. We also know that the torque acting on this element is
\begin{align*}
\mathrm{d}\tau &= r\,\mathrm{d}F_t\\
&= ra_t\,\mathrm{d}m
\end{align*}

Equation {eq}`eq:C9_rolla` defines the relationship between the tangential and angular accelerations, and thus
\begin{align*}
\mathrm{d}\tau &= r^2\alpha\,\mathrm{d}m
\end{align*}
To find the net torque we integrate over the entire mass and make use of the fact that the angular acceleration is the same for all mass elements if the body is rigid. Therefore
\begin{align*}
\tau &= \alpha\int r^2\,\mathrm{d}m\\
&= I\alpha
\end{align*}

### A return to the pendulum
Do you remember way back in [Lecture 2: the pendulum](sec:C2_pend) where we derived the differential equation (equation {eq}`eq:C2_simple`) for a pendulum using the components on the gravitational force acting along the path of the pendulum? A pendulum is a rotating system though, albeit one that never completes a full rotation[^pend], but we can still use our rotational dynamics and our new understanding of torque to find the same result.
[^pend]: Unless you **really** push the pendulum. But then we are definitely away from the small angle domain...

Starting from our definitions of torque
\begin{align*}
\boldsymbol{\tau} = \textbf{r}\times\textbf{F} = \frac{\mathrm{d}\textbf{L}}{\mathrm{d}t}
\end{align*}
We can define the torque acting on the pendulum in terms of the length and force due to gravity vectors as
\begin{align*}
\boldsymbol{\tau} &= \textbf{L}\times\textbf{F}_g\\
\therefore |\boldsymbol{\tau}| &= -mgL\sin\theta
\end{align*}
Next the angular momentum can be written as
\begin{align*}
\textbf{L} &= \textbf{r}\times\textbf{p}\\
	&= m\textbf{r}\times\left(\boldsymbol{\omega}\times\textbf{r}\right)\\
\therefore |\textbf{L}| &= mL^2\omega\\
	&= mL^2\frac{\mathrm{d}\theta}{\mathrm{d}t}\\
	\text{and} \qquad\frac{\mathrm{d}|\textbf{L}|}{\mathrm{d}t} &= mL^2\frac{\mathrm{d}^2\theta}{\mathrm{d}t^2}
	\end{align*}
If $\boldsymbol{\tau} = \dfrac{\mathrm{d}\textbf{L}}{\mathrm{d}t}$ then $|\boldsymbol{\tau}| = \dfrac{\mathrm{d}|\textbf{L}|}{\mathrm{d}t}$ which gives
\begin{align*}
-mgL\sin\theta &= mL^2\frac{\mathrm{d}^2\theta}{\mathrm{d}t^2}\\
\text{or}\qquad \frac{\mathrm{d}^2\theta}{\mathrm{d}t^2}+\frac{g}{L}\sin\theta &= 0
\end{align*}
which is exactly the same as equation {eq}`eq:C2_simple` we derived earlier.

## Angular momentum
### Definition of angular momentum
To find the total angular momentum of a solid rotating body we once again consider it to be made of small mass elements $\mathrm{d}m$ each some distance $r_i$ and moving with some momentum $p_i$. The angular momentum for each element $L_i$ is
\begin{align*}
L_i &= \textbf{r}_i\times \textbf{p}_i\\
&= v_ir_i\,\mathrm{d}m\\
&= r_i^2\omega\,\mathrm{d}m
\end{align*}

````{margin}
```{note}
The choice of $z$ here is entirely arbitrary, but defining an axis helps us to remember that rotational systems need the axis of rotation stating
```
````
So the total angular momentum of the body rotating about the $z$axis is simply the sum of the angular momenta for the individual mass elements, i.e.
```{math}
:label: eq:C10_L
L_z &= \omega\int r_i^2\,\mathrm{d}m\nonumber\\
&= I\omega
```

Often we are more concerned with how the momentum is changing, so by taking the time derivative of equation {eq}`eq:C10_L` we find
\begin{align*}
\frac{\mathrm{d}L_z}{\mathrm{d}t} &= I\frac{\mathrm{d}\omega}{\mathrm{d}t}\\
&= I\alpha\\
\therefore \frac{\mathrm{d}L_z}{\mathrm{d}t} &= \tau
\end{align*}

This final equation should come as no surprise. Back in [Lecture 3](Lecture3) we saw the more general form of Newton's Second Law (equation {eq}`eq:C3_Second_new`) which states that the rate of change of linear momentum is equal to the net force applied. Here we have found a rotational equivalent between the rate of change of angular momentum and the net applied torque.

### Conservation of angular moment
````{margin}
```{note}
Another shout-out to Noether's Theorem here for the proof that angular momentum **must** be a conserved quantity
```
````
If the net external torque acting on a rotating system is zero then the total angular momentum is constant. This is very similar to the conservation of linear momentum case but extra care must be taken when working with angular momentum as we must consider the distribution of the mass (due to the moment of inertia term in equation {eq}`eq:C10_L`).

There are some really good physical examples of this, one of which I hope you are all familiar with but if not you can find some good videos on YouTube. The system I am thinking of is an ice-skater who is performing a spin on the spot. When they are spinning with their arms outstretched they are rotating with some angular velocity but if they bring their arms in suddenly their rotational velocity increase. We assume they have not suddenly changed their mass and we can see there are no external forces acting on them. The cause of this increased angular velocity is their change in mass distribution.

Consider the simplified equation for conservation of angular momentum for this system.
\begin{align*}
L_\text{before} &= L_\text{after}\\
I_\text{before}\omega_\text{before} &= I_\text{after}\omega_\text{after}
\end{align*}
````{margin}
```{note}
Remember moment of inertia is mass times distance squared, so bringing arms in reduces the distance for some of the masses.
```
````
If the ice skater decreases their moment of inertia $I_\text{after}$ then their angular velocity $\omega_\text{after}$ must increase by an equal factor so that the conservation of angular momentum equation above is balanced.

`````{admonition} Question 1
```{tab-item} Question
Suppose that while opening a 1.0 m wide door you push against the edge farthest from the hinge, applying a force with constant magnitude $0.90$ N at right angles to the surface of the door. How much work do you do on the door during an angular displacement of $\Delta\phi=30^{\circ}$?
```

```{tab-item} Hint
A door is equivalent to a uniform rod, at least in terms of the moment of inertia. Just be careful to ensure you've got the right axis of rotation location.
```

```{tab-item} Solution
For a constant torque the work done is $W=\tau \Delta\phi$ and the torque is
\begin{align*}
\tau &= FR\sin\theta\\
	&= FR
\end{align*} 
where $\theta$ is the angle between the applied force and the length vectors. Therefore
\begin{align*}
W &= FR\Delta\phi\\
		&= 0.90 \times 1.0 \times \frac{30\times\pi}{180}\\
		&= 0.47\text{ J}
\end{align*}
```
`````

`````{admonition} Question 3
```{tab-item} Question
The rotor of the gyroscope on the Gravity Probe B experiment is a quartz sphere of diameter $3.8\text{ cm}$ and mass $7.61\times10^{-2}\text{ kg}$. To start the sphere spinning, a stream of helium gas flowing into an equatorial channel in the surface of the housing is blown tangentially against the rotor. 
1. What torque must this stream of gas exert on the rotor to accelerate it uniformly from $0$ to $10000$ rpm in 30 minutes?
2. What force must it exert on the equator of the sphere?
```

```{tab-item} Hint
If you struggle with torque and angular acceleration then consider a linear analogue with force and acceleration.

For part 2 you should draw out the system and include the force vectors, from which you can identify any angles between vectors and hopefully simplify the system.
```

```{tab-item} Solution
**1**
The final angular velocity $\omega_f=1.05\times10^3 \text{ radians s}^{-1}$ and so the angular acceleration is
\begin{align*}
\alpha &= \frac{\omega_f - \omega_i}{t_f - t_i}\\ 
&= \frac{1.05\times10^3 - 0}{30\times60 - 0}\\
&= 0.582\text{ radians s}^{-2}
\end{align*}
The required torque is therefore
\begin{align*}
\tau &= I\alpha\\
&= \frac{2}{5}MR^2\times\frac{\omega_f}{t_f}\\
&= 6.4\times10^{-6}\text{ N m}
\end{align*}
        
**2**
The gas is blown tangentially against the rotor, such that the angle between the force from the gas and the radius vector of the sphere is $90^{\circ}$.
\begin{align*}
\tau &= FR\sin\theta = FR\\
\therefore F &= \frac{\tau}{R}\\
&= 3.4\times10^{-4}\text{ N}
\end{align*}
`````
